```python
correlation_matrix = data.corr()



# Crea la figura y los ejes de subgráficos

fig, ax = plt.subplots(figsize=(10, 8))



# Mapa de calor para la matriz de correlación (triangulo superior)

sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm',

      cbar=False, square=True, ax=ax, linewidths=0.5)

for i in range(len(correlation_matrix.columns)):

  for j in range(i):

    ax.text(j+0.5, i+0.5, f'{correlation_matrix.values[i,j]:.2f}',

        ha='center', va='center', color='black')



# Añade los gráficos de dispersión en el triángulo inferior

for i in range(len(correlation_matrix.columns)):

  for j in range(i):

    ax = plt.subplot(len(correlation_matrix.columns), len(correlation_matrix.columns),

             i * len(correlation_matrix.columns) + j + 1)

    plt.scatter(data.iloc[:, j], data.iloc[:, i], color='blue', alpha=0.5)

    plt.xlabel(data.columns[j])

    plt.ylabel(data.columns[i])
# Añade las distribuciones en la diagonal

for i in range(len(correlation_matrix.columns)):

  ax = plt.subplot(len(correlation_matrix.columns), len(correlation_matrix.columns),

           i * len(correlation_matrix.columns) + i + 1)

  sns.histplot(data.iloc[:, i], kde=True, color='blue', ax=ax)

  plt.xlabel(data.columns[i])



plt.tight_layout()

plt.show()
```

```python
#GRAFICO 2: es un mapa de calor con la correlación de las variables entre si

# Calcula la matriz de correlación

correlation_matrix = data.corr()



# Crea el gráfico de matriz de correlación

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)

plt.title('Matriz de Correlación')

plt.show()
```

Gracias a estos dos gráficos, podemos observar que el dataset cuenta con poca correlación entre las variables, con excepción de un par. esto es positivo a la hora de analizar la información, ya que variables correlacionadas implicarían redundancia, que preferiríamos no tener.
las variables con mayor correlación son Width - Length y Length - Heigh, que se encuentran en un nivel amarillo/naranja. el resto tiende al azul .

Mi recomendación sería eliminar alguna de las variables (Width o Length) para disminuir la cantidad de columnas del dataset.

El propósito es discernir el impacto de estos atributos en la probabilidad de presentar reclamos y determinar los factores más influyentes. Además, se investiga la suposición de que ciertos dispositivos electrónicos o mecánicos integrados en los vehículos podrían influir en la frecuencia de reclamaciones recibidas.

Al lograr este objetivo, se espera mejorar significativamente la gestión del riesgo y perfeccionar el proceso de determinación de primas de seguros, aportando una mayor precisión en el cálculo de riesgos.

Lo que podemos observar en nuestro dataset es información acerca de asegurados, como por ejemplo: antigüedad del vehículo, edad del conductor, modelo del vehículo, motor, etc, y finalmente la variable objetivo que indica si el asegurado inició un reclamo. Con esta información, podremos ver cuales son los atributos de aquellas personas que realizaron un reclamo, o cuales son las variables que mejor explican el comportamiento de la variable objetivo.
Las preguntas que se pueden realizar para este caso de estudio son las siguientes:
-	Existe alguna correlación entre los atributos de los asegurados, tales como antigüedad, edad, modelo, etc, y la probabilidad de ocurrencia del siniestro?
-	De ser así, cuales serían las variables más influyentes de mayor a menor?
-	Existe alguna variable que no tenga influencia en absoluto?

Hipótesis:
Variables que explican la tenencia o no de algún dispositivo electrónico/mecánico deberían tener influencia en la cantidad de reclamos recibidos.
Por ejemplo, “transmisión_type” es una variable que indica si el motor tiene transmisión manual o automática. Uno podría pensar que los vehículos de transmisión automática son más simples de utilizar, y en consecuencia deberían tener menor cantidad de reclamos. Lo mismo ocurriría con variables como “is_parking_sensors”, “is_parking_camera”, “rear brakes_type”, entre otros.
Sin embargo, pareciera que estas variables no tienen influencia, debido a las visualizaciones que se desarrollan a continuación.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', 44)
```

```python
siniestros = pd.read_csv("https://raw.githubusercontent.com/danielcaba889/Car-insurance-claim-prediction/main/Car%20insurance%20claims.csv")
siniestros.head()
```

Observamos que algunos de los atributos de nuestros dataset son numéricos y otros categóricos. Los categóricos podrían generar una problemática a la hora de construir un modelo, por lo tanto los elimino. Me quedo con las variables numéricas y los de clasificacion de tipo "YES/NO"

```python
siniestros.dtypes
```

```python
#Dropeo la variable target
data = siniestros.drop('is_claim', axis = 1)
target = siniestros['is_claim']
```

```python
#Elimino variables categóricas
data.drop(['policy_id',
           'area_cluster',
           'segment', 'model',
           'fuel_type',
           'max_torque',
           'max_power',
           'engine_type',
           'rear_brakes_type',
           'transmission_type',
           'steering_type'], axis = 1, inplace = True)
```

```python
#selecciono variables YES/NO
yes_no_columns = []

for col in data.columns:
    if data[col].isin(['Yes', 'No']).any():
        yes_no_columns.append(col)

print("Columnas con valores 'Yes' o 'No':")
print(yes_no_columns)
```

```python
columnas_yesno = ['is_esc',
 'is_adjustable_steering',
 'is_tpms',
 'is_parking_sensors',
 'is_parking_camera',
 'is_front_fog_lights',
 'is_rear_window_wiper',
 'is_rear_window_washer',
 'is_rear_window_defogger',
 'is_brake_assist',
 'is_power_door_locks',
 'is_central_locking',
 'is_power_steering',
 'is_driver_seat_height_adjustable',
 'is_day_night_rear_view_mirror',
 'is_ecw',
 'is_speed_alert']
```

```python
data.head()
```

```python
#Convierto YES en 1 (True), NO en 0 (False)
true_false_columns = data[columnas_yesno] == "Yes"
true_false_columns.columns
```

```python
data_num_values =  pd.concat([data.drop(true_false_columns.columns, axis = 1), true_false_columns], axis=1)
data_num_values.head()
```

```python
#Escalo los datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data_num_values)
```

```python
scaled_data = scaler.transform(data_num_values)
```

## Principal components analysis (PCA)

###elijo un método de feature selection para reducir la dimensionalidad del dataset


```python
from sklearn.decomposition import PCA
```

```python
pca = PCA (n_components = 2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
print(scaled_data.shape)
print(x_pca.shape)
```

```python
# Al reducir la dimensionalidad del dataset (de 32 a 2), se puede graficar en el eje X e Y y coloreamos la variable target para distinguir del resto.
#Se observa un gráfico muy interesante

plt.figure(figsize = (8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = target)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#divido el dataset en 4. con X_train entreno el modelo, teniendo como variable objetivo y_train
#con X_test e y_test realizaré las predicciones y evaluaré la performance del modelo

X_train, X_test, y_train, y_test = train_test_split( scaled_data, target, test_size = 0.3, random_state = 889 )
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=889)
clf.fit(X_train, y_train)
```

```python
from sklearn import tree
tree.plot_tree(clf)
plt.show()
```

El modelo de árboles de decisión tiene demasiado overfitting. Esto se puede observar al ver la cantidad de separaciones que intenta hacer el modelo, tratando de llegar al valor correcto por medio de una exagerada cantidad de separaciones.

```python
y_train_predict = clf.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_true = y_train, y_pred = y_train_predict)
```

```python
y_test_predict = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_true = y_test, y_pred = y_test_predict)
```

Lo que podemos observar es que el modelo "memoriza" los valores entrenados con el set de entrenamiento (X_train) y realiza predicciones perfectas, al contrastarse con la variable target (y_train).
Sin embargo, al correr el modelo en el set de prueba, observamos que las predicciones no son tan buenas. Esto se debe a que el modelo tiene mucho overfitting.



Una posible mejora podría ser disminuir la cantidad de hojas que tiene el modelo

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 15, random_state=889)
clf.fit(X_train, y_train)
```

```python
max_depth_y_train_predict = clf.predict(X_train)
confusion_matrix (y_pred = max_depth_y_train_predict, y_true = y_train)
```

```python
max_depth_y_test_predict = clf.predict(X_test)
confusion_matrix (y_pred = max_depth_y_test_predict, y_true = y_test)
```

Se observa que al utilizar el parámetro max_depth evaluado en 15, reducimos el overfitting. Sin embargo, esto no mejora la calidad predictiva en el set de testeo.

El modelo evaluado en el set de testeo arroja los siguientes resultados:

32 casos positivos correctamente clasificados
